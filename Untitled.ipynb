```python
%matplotlib inline
from matplotlib import pyplot
plt.pyplot.hist(titanic["Age"])

# set x/y labels and plot title
plt.pyplot.xlabel("Age")
plt.pyplot.ylabel("total")
plt.pyplot.title("age bins")
```

```python
import numpy as np

bins = np.linspace(min(titanic["Age"]), max(titanic["Age"]), 6)
bins
```

```python
group_names = ['muda', 'remaja', 'dewasa', 'tua', 'lansia']
```

```python
titanic['Age-binned'] = pd.cut(titanic['Age'], bins, labels=group_names, include_lowest=True )
titanic[['Age','Age-binned']].head(20)
```

```python
# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(titanic["Age"], bins = 6)

# set x/y labels and plot title
plt.pyplot.xlabel("age")
plt.pyplot.ylabel("total")
plt.pyplot.title("age bins")
```
